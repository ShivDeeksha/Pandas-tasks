### Creating a Subset of the Kindle Data
Select specific columns from the DataFrame to create a new subset.


```python
kindle_subset=kindle[["title","author","price","stars","soldBy"]]
kindle_subset.head()
```

### Filtering Books with High Ratings
Create a subset of the Kindle data for books with stars greater than 4.5.


```python
best_books = kindle_subset[kindle_subset["stars"]>4.5]
best_books.head()
```

### Filtering Books Sold by Amazon with High Ratings
Create a subset for books sold by Amazon with stars greater than 4.5.


```python
sold_by_amazon= kindle_subset[(kindle_subset["soldBy"].isin(["Amazon.com Services LLC"])) & (kindle_subset["stars"]>4.5)]
sold_by_amazon.head()
```

### Filtering Books with Non-Null Sellers and High Ratings
Create a subset for books with a non-null "soldBy" field and stars greater than 4.5.

```python
sold_by_nonull= kindle_subset[(kindle_subset["soldBy"].notna()) & (kindle_subset["stars"]>4.5)]
sold_by_nonull.head()
```

### Extracting Authors with Perfect Ratings
Filter the authors of books that have a 5-star rating.

```python
best_authors = kindle.loc[kindle["stars"]==5,"author"]
best_authors.head(10)
```

### Extracting a Subset Using `iloc`
Use `iloc` to select a specific range of rows and columns from the DataFrame.


```python
best_authors = kindle.iloc[10:22,2:5]
best_authors.head()
```
